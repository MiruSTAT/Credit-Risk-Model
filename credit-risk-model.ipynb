### **Introduction**

In this Project I will be attempting to build my first credit risk model using the [Give Me Some Credit dataset](https://www.kaggle.com/datasets/brycecf/give-me-some-credit-dataset). At its core, a credit risk model is used to predict the probability that the borrower doesn't pay back an organization, Bank, etc. when lended an amount of money

**<u>What is Default?</u>**  
Default is a term used to identify borrowers who fail to pay back in a fixed amount of time, usually 90+ days or 90DPD (90 days past due)

**<u>Core components</u>**
1. Probability of Default [PD] --> probability of borrower not paying back
$$PD = P(Borrower Defaults)$$
2. Loss Given Default [LGD] --> if borrower doesn't pay back, how much money did we lose?
$$LGD = \frac{Loss After Recovery}{Total Exposure}$$
3. Exposure at Default [EAD] --> if borrower doesn't pay back, how much money is at risk at a given time period?
$$EAD = \text{Balance at time of Default}$$
4. Expected Loss [EL] --> the average loss 
$$EL = PD\times LGD\times EAD$$

**<u>Model Types</u>**  
- Application - Risk at time of Application(Static)
- Behavioral - Monitor ongoing Customer behavior
- Collections - Predict repayment for delinquent customers
- Regulatory - Capital Adequecy & Loss Allowance
